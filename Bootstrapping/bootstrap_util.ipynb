```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
FLs = pd.read_csv('Bootstrap/FLs.csv') 
AFBs = pd.read_csv('Bootstrap/AFBs.csv') 
S3s = pd.read_csv('Bootstrap/S3s.csv') 
S4s = pd.read_csv('Bootstrap/S4s.csv') 
S5s = pd.read_csv('Bootstrap/S5s.csv') 
S7s = pd.read_csv('Bootstrap/S7s.csv') 
S8s = pd.read_csv('Bootstrap/S8s.csv') 
S9s = pd.read_csv('Bootstrap/S9s.csv') 
```

```python
files = [FLs,AFBs,S3s,S4s,S5s,S7s,S8s,S9s]
files_string = ['FLs','AFBs','S3s','S4s','S5s','S7s','S8s','S9s']
bin_names = ['Bin'+str(i) for i in range(10)]
```

```python
for file, string_file in zip(files,files_string):
      for bin_name in bin_names:
            bin_data = file[bin_name]
            median = np.percentile(bin_data, 50)

            length = len(bin_data)

            # get 95% interval
            alpha = 100-95
            lower_ci = np.percentile(bin_data, alpha/2)
            upper_ci = np.percentile(bin_data, 100-alpha/2)

            print(f"Model accuracy is reported on the {string_file} {bin_name}. {length} bootstrapped samples " 
                  f"were used to calculate 95% confidence intervals.\n"
                  f"Median accuracy is {median:.2f} with a 95% a confidence "
                  f"interval of [{lower_ci:.2f},{upper_ci:.2f}].")

            fig = plt.figure(figsize=(12,8))
            sns.kdeplot(bin_data)
            plt.title(f"{string_file} {bin_name} across {length} bootstrap samples of ML Signal\n"
                      f"showing median with 95\\% confidence intervals [{lower_ci:.2f},{upper_ci:.2f}]")
            plt.xlabel("Accuracy")
            plt.axvline(median,0, 14, linestyle="--", color="red")
            plt.axvline(lower_ci,0, 14, linestyle="--", color="red")
            plt.axvline(upper_ci,0, 14, linestyle="--", color="red")
            plt.savefig(f'Bootstrap_Graphs/{string_file}_{bin_name}.jpg',dpi=600)
```

```python
def addToFile(file, msg):
    f = open(file, 'a').write(msg+'\n') 
```
