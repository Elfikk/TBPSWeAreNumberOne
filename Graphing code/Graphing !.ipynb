```python
import pandas as pd
from matplotlib import pyplot as pl
```

```python
#Load in the files you need here
data1 = pd.read_csv('BackgroundData/signal.csv', on_bad_lines='skip')
data2 = pd.read_csv('BackgroundData/total_dataset.csv', on_bad_lines='skip')
data3 = pd.read_csv('BackgroundData/total_dataset.csv', on_bad_lines='skip')
#Fit = pd.read_csv('BackgroundData/total_dataset.csv', on_bad_lines='skip')

#If you did so, and only need a specific column, please choose it here
#X axis goes here
DataX1 = data1["Unnamed: 0.1"]
DataX2 = data2["Unnamed: 0.1"]
DataX3 = data3["Unnamed: 0.1"]
FitX = 1

#Y axis goes here
DataY1 = data1["B0_M"]
DataY2 = data2["q2"]
DataY3 = data3["Kstar_M"]
FitY = 1
```

```python
#If you just need a simple plot, here it is !
pl.plot(DataX1, DataY1, ".", color = "Teal", label = "Plot 1")
pl.plot(DataX2, DataY2, ".", color = "Crimson", label = "Plot 2")
pl.plot(DataX3, DataY3, ".", color = "Chartreuse", label = "Plot 3")

#THIS LINE IS ONLY FOR FITTING !!!
#pl.plot(FitX, FitY, ".", color = "Chartreuse", label = "Plot 3")

#pl.errorbar(x,y, yerr = yerr , ls = 'none', color = 'black')
pl.xlabel("X Title")
pl.ylabel("Y Title")
pl.title("Hey, here is a title !")
pl.legend()
pl.plot()

pl.savefig("Graph.png")
```
